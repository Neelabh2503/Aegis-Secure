```python
import pandas as pd
import seaborn as sns
from collections import Counter
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data_new.csv')
df=df.iloc[1:15000,:]
df.head()
df.info()
df=df.dropna()
df.drop_duplicates(inplace=True)
```

```python
df['message_length'] = df['Message'].astype(str).str.len()
print(df['message_length'].mean())
print(df['message_length'].mode())
print(df['message_length'].median())
sns.histplot(df['message_length'], bins=300, kde=True)
```

```python
sns.histplot(df['label'])
```

```python
all_words = " ".join(df['Message'].astype(str)).lower().split()
common_words = Counter(all_words).most_common(25)
words, counts = zip(*common_words)
plt.barh(words, counts)
plt.gca().invert_yaxis()
plt.show()
```

```python
sns.scatterplot(x='message_length', y='label', data=df, alpha=0.3)
plt.xlabel("Message Length)")
plt.ylabel("Label")
plt.title("Scatter Plot of Message Length vs Label")
plt.show()
```
